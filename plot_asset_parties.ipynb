```python
fig = plt.figure(figsize=(8, 8), dpi=100)
ax = fig.add_subplot(111)

year = np.arange(2007, 2022)

tick_num = 4
legend_fontsize=15
tick_labelsize=20
ax.tick_params('both', which='major', labelsize=tick_labelsize, length=7, width=1, direction='in', top=True, right=True)
ax.tick_params('both', which='minor',labelsize=tick_labelsize,  length=3, width=1, direction='in', top=True, right=True)

colorlist = ["#cb5242",
             "#60b851",
            "#b45ac2",
            "#b3b245",
            "#6f7dcb",
            "#db8f46",
            "#4eb9a5",
            "#548341",
            "#c55d88",
            "#8e7335"]

lw_line = 3
lw_scatter= 4

start = 4
end = 15

year =year[start: end]

plt.scatter(year, TMC[start: end]/TMC[start: end][0], lw=5, color = colorlist[0], label = "TMC")
plt.plot(year, TMC[start: end]/TMC[start: end][0], lw= 3, color = colorlist[0])

plt.scatter(year, BJP[start: end]/BJP[start: end][0], lw=lw_scatter, color = colorlist[2], label = "BJP")
plt.plot(year, BJP[start: end]/BJP[start: end][0], lw= lw_line, color = colorlist[2])

plt.scatter(year[1:], BSP[start: end]/BSP[start: end][0], lw=lw_scatter, color = colorlist[1], label = "BSP")
plt.plot(year[1:], BSP[start: end]/BSP[start: end][0], lw= lw_line, color = colorlist[1])


plt.scatter(year, CPI[start: end]/CPI[start: end][0], lw=lw_scatter, color = colorlist[3],  label = "CPI")
plt.plot(year, CPI[start: end]/CPI[start: end][0], lw= lw_line, color = colorlist[3])

plt.scatter(year, CPIM[start: end]/CPIM[start: end][0], lw=lw_scatter, color = colorlist[4], label = "CPI(M)")
plt.plot(year, CPIM[start: end]/CPIM[start: end][0], lw= lw_line, color = colorlist[4])

plt.scatter(year, INC[start: end]/ INC[start: end][0], lw=lw_scatter, color = colorlist[5], label = "INC")
plt.plot(year, INC[start: end]/ INC[start: end][0], lw= lw_line, color = colorlist[5])

plt.scatter(year, NCP[start: end]/ NCP[start: end][0], lw=lw_scatter, color = colorlist[6],  label = "NCP")
plt.plot(year, NCP[start: end]/ NCP[start: end][0], lw= lw_line, color = colorlist[6])



ax.set_xlabel(r'Year' , labelpad = 15)
ax.set_ylabel(r'Asset multiplication factor (since 2011)', labelpad = 15)
ax.set_title(r'Asset increament rate of Indian political parties', fontsize= 20 , pad = 15)

ax.set_xlim(2011, 2022)
ax.set_ylim(0, 43)

ax.axvspan(2018, 2021, alpha=.1, color='red')

legend_fontsize = 20
ax.legend(loc='upper left', fontsize=legend_fontsize, handlelength=1,
             frameon=False, framealpha=0.0, labelspacing=.3,
               handletextpad=0.7, borderpad=0.55, scatterpoints=1)
#plt.yscale("")

plt.savefig("increament.jpg", bbox_inch = "tight")
```
